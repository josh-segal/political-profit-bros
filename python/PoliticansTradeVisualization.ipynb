```python
import pandas as pd 
import numpy as np
import plotly.express as px 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('politician_dataset.csv')
df
```

```python
buy_df = df[df['Type'] == 'Buy']
companies_grouped = buy_df.groupby('Ticker').sum().sort_values(by='Trade Value', ascending = False).reset_index()
top_ten_companies = companies_grouped.head(10)

```

```python
plt.bar(top_ten_companies["Ticker"], top_ten_companies['Trade Value'], color="Green")
plt.title("Top 10, most Bought Companies by Congress in a Year")
plt.xlabel("Companies")
plt.ylabel("Value (in millions)")
plt.show()
```

```python
sell_df = df[df['Type'] == 'Sell']
sold_companies = sell_df.groupby('Ticker').sum().sort_values(by='Trade Value', ascending = False).reset_index().head(10)
sold_companies
```

```python
plt.bar(sold_companies["Ticker"], sold_companies['Trade Value'], color="Red")
plt.title("Top 10, most Sold Companies by Congress in a Year")
plt.xlabel("Companies")
plt.ylabel("Value (in millions)")
plt.show()
```
